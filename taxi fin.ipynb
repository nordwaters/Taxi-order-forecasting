Компания Noname собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

Рассчитаем признак lag при помощи seasonal_decompose.

```python
decomposed_data_24 = seasonal_decompose(data['2018-08-30':'2018-08-31'])
max_lag = ((decomposed_data_24.seasonal.idxmax() - decomposed_data_24.seasonal.idxmin()) 
            // np.timedelta64(1, 'h'))
max_lag
```

Видно, что самые популярные дни по заказам такси - это понедельник и пятница, реже всего заказывают такси во вторник и воскресенье. Добавим в датасет столбцы, которые можно будет использовать в качестве признаков.

## Обучение

### Создание обучающей, валидационной и тестовой выборок

```python
train, test = train_test_split(data, shuffle = False, random_state = RANDOM_STATE, test_size = 0.2)
valid, test = train_test_split(test, shuffle = False, random_state = RANDOM_STATE, test_size = 0.5)
train
```

```python
# Проверим результаты

print('Размер тренировочного датасета - {:.0%}'.format(
    train.shape[0]/data.shape[0]))
print('Размер валидационного датасета - {:.0%}'.format(
    valid.shape[0]/data.shape[0]))
print('Размер тестового датасета - {:.0%}'.format(
    test.shape[0]/data.shape[0]))
```

```python
# определим признаки и таргеты

train_features = train.drop('num_orders', axis=1)
train_target = train['num_orders']

valid_features = valid.drop('num_orders', axis=1)
valid_target = valid['num_orders']

test_features = test.drop('num_orders', axis=1)
test_target = test['num_orders']
```

Для расчёта целевой метрики будет использоваться RMSE. Создадим функцию для расчёта.

```python
def rmse(test, predict):
    return ((mean_squared_error(test, predict))**0.5)
```

### TimeSeriesSplit

```python
# Создадим переменную со значением для TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=3)
```

### Линейная регрессия

```python
# Инициализация модели линейной регрессии

model_lr=LinearRegression()

model_lr.fit(train_features,train_target)
predict_lr = model_lr.predict(valid_features)


mae_lr = mean_absolute_error(valid_target, predict_lr)
print('MAE для Линейной Регрессии:', mae_lr)

rmse_lr = rmse(valid_target, predict_lr)
print('RMSE для Линейной Регрессии:', rmse_lr)
```

### Случайный лес

```python
# Инициализация модели случайного леса

parameters = {'n_estimators' : [5, 10, 20, 30, 50],
              'max_depth'    : [4, 6, 8, 10, 20]
}


model_rf = RandomForestRegressor(random_state=RANDOM_STATE)
grid_rf = GridSearchCV(model_rf, parameters, scoring='neg_root_mean_squared_error', cv=tscv)
grid_rf.fit(train_features, train_target)

best_score_rf = np.abs(grid_rf.best_score_)

print('Значение RMSE на обучающей выборке равно:', best_score_rf)
print('Лучшие параметры:', grid_rf.best_params_)
```

Проверим модель Случайного леса на валидационной выборке.

```python
predict_rf=grid_rf.predict(valid_features)

mae_rf = mean_absolute_error(valid_target, predict_rf)
print('MAE модели Случайный лес равна:', mae_rf)

rmse_rf=rmse(valid_target, predict_rf)
print('RMSE модели Случайный лес равна:', rmse_rf)
```

### Категориальный бустинг

```python
model_cbr = CatBoostRegressor(random_state=RANDOM_STATE,learning_rate=0.2,
                              verbose=100, n_estimators=300)
model_cbr.fit(train_features, train_target, eval_set=(valid_features, valid_target))
```

CatBoost на валидационной выборке.

```python
predict_cbr = model_cbr.predict(valid_features)

mae_cbr = mean_absolute_error(valid_target, predict_cbr)
print('MAE модели Категориального бустинга равна:', mae_cbr)

rmse_cbr=rmse(valid_target, predict_cbr)
print('RMSE модели Категориального бустинга равна:', rmse_cbr)
```
